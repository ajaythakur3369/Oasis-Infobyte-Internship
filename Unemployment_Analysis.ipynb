```python
# Sample DataFrame
data = {'Region': ['A', 'B', 'C', 'D'],
        'Unemployment Rate': [5.2, 6.0, 4.8, 7.2]}  # Example data, replace with your actual data
df = pd.DataFrame(data)

# Plotting
plt.bar(df['Region'], df['Unemployment Rate'])
plt.xticks(rotation=90)
plt.xlabel('States')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment according to States')
plt.show()
```

```python
# Sample DataFrame
data = {'Estimated Unemployment Rate (%)': [5.2, 6.0, 4.8, 7.2], 'Area': ['A', 'B', 'A', 'B']}
df = pd.DataFrame(data)

# Plotting
sns.histplot(x='Estimated Unemployment Rate (%)', data=df, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```
